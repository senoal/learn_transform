```python
#mencari suara sah x = c1+c2+c3-bb

df['suara_sah'] = df.iloc[:, 1:4].sum(axis=1)
df
```

```python
#mencari suara hilang x = total_pop - suara_sah

df['suara_hilang'] = df['Jml_wrg']-df['suara_sah']
df
```

```python
#mencari jumlah total per kolom
# step 1, ubah kolom menjadi baris
df=df.T
df
```

```python
#step 2, tambah kolom dengan penjumlahan yang di inginkan

df['JML'] = df.iloc[1:, 0:15].sum(axis=1)
df
```

```python
#putar tabel untuk melihat hasil

df=df.T
df
```
